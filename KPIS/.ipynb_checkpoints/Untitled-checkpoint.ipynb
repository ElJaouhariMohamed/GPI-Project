```python
import pandas as pd

def calculate_kpis(file_path, temps_base):
    # Charger le fichier CSV dans un DataFrame
    df = pd.read_csv(file_path)
    
    # Calculer le temps brut de fonctionnement (en heures)
    temps_brut = df['failure'] + df['realFail']
    
    # Calculer le temps net de fonctionnement (en heures)
    temps_maintenance_predictive = df['realFail']
    temps_net = temps_base + temps_maintenance_predictive
    
    # Calculer la productivité
    productivite = temps_net / temps_brut * 100
    
    # Ajouter les colonnes des KPIs au DataFrame
    df['temps_brut'] = temps_brut
    df['temps_net'] = temps_net
    df['productivite'] = productivite
    
    # Enregistrer le nouveau fichier CSV avec les nouvelles colonnes
    new_file_path = file_path.split('.csv')[0] + '_with_kpis.csv'
    df.to_csv(new_file_path, index=False)
    
    # Retourner le DataFrame avec les nouvelles colonnes
    return df
```

les kpis MTBF et MTTR

```python
import pandas as pd

def calculate_kpis(file_path, temps_base):
    # Charger le fichier CSV dans un DataFrame
    df = pd.read_csv(file_path)
    
    # Calculer le temps brut de fonctionnement par heures
    temps_brut = df['failure'] + df['realFail']
    
    # Calculer le temps net de fonctionnement oar heures
    temps_maintenance_predictive = df['realFail']
    temps_net = temps_base + temps_maintenance_predictive
    
    # Calculer la productivité
    productivite = temps_net / temps_brut * 100
    
    # MTBF en heures    
    temps_entre_pannes = temps_brut.diff().shift(-1)
    temps_entre_pannes.iloc[-1] = temps_brut.iloc[-1]
    mtbf = temps_entre_pannes.mean()
    
    # MTTR en heures
    mttr = df['realFail'].mean()
    
    # Ajouter les colonnes des KPIs au DataFrame
    df['temps_brut'] = temps_brut
    df['temps_net'] = temps_net
    df['productivite'] = productivite
    df['mtbf'] = mtbf
    df['mttr'] = mttr
    
    # Enregistrer le nouveau fichier CSV avec les nouvelles colonnes
    new_file_path = file_path.split('.csv')[0] + '_with_kpis2.csv'
    df.to_csv(new_file_path, index=False)
    
    # Retourner le DataFrame avec les nouvelles colonnes
    return df
```

```python
df_with_kpis2 = calculate_kpis('results.csv', 6)
```
